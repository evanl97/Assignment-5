```python
import glob
import numpy as np
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt

dataset = sorted(glob.glob('data*.csv'))
```

```python
def identify_issues(filename):
    """Determine if the dataset has any major issues or not"""
    data = pd.read_csv(filepath_or_buffer = filename)
    max_data = data.max()
    if max_data[0] == 0 and max_data[20] == 20:
        print("I doubt the integrity of this dataset")
    else:
        min_data = data.min()
        if sum(min_data[:]) == 0:
            print("Why is this happening?")
        else:
            print("Dataset looks good")
```

```python
def visualize(filename):
    "Visualize the subplots of the 4 datasets"
    data = pd.read_csv(filepath_or_buffer = filename)
    data.mean().plot(kind='line', label = 'Average')
    data.max().plot(kind='line', label = 'Max')
    data.min().plot(kind='line', label = 'Min')
    plt.legend()
    plt.xlabel("feature #")
    plt.ylabel("values")
    plt.show()
```

```python
for datasets in dataset[:4]:
    print("\n")
    print (datasets)
    visualize(datasets)
    identify_issues(datasets)
```
